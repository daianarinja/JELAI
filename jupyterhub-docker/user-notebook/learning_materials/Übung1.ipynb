### 1.1.3 Entfernen von Studienabbrecher*innen

```r
AllFollowingMissing <- function(row){
  Missings <- which(is.na(row))
  if(length(Missings)==0) return(FALSE)
  if(length(Missings) < length(row)){
    all(is.na(row[(Missings[length(Missings)]):length(row)]))
  } else TRUE
}
RowsWithAllFollowingMissing <- apply(data,1,AllFollowingMissing)
RowsToOmit <- which(RowsWithAllFollowingMissing)
data <- data[-RowsToOmit,]
```
